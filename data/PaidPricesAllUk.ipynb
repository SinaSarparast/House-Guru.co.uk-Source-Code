```python
frames = []
for i in range(1995,2018):
#     print('../PaidPrice/pp-'+str(i)+'.csv')
    frames.append(pd.read_csv('../data/PaidPrice/pp-'+str(i)+'.csv',encoding='latin-1',header=None,usecols=[1,2,3],names=['price','date','postcode']))

```

```python
paidprices = pd.concat(frames)
```

```python
postcodes = pd.read_csv('2018_postcode_mapping.csv',usecols=['pcds','lat','long'])
```

```python
postcodes = postcodes.rename(index=str,columns={"pcds":"postcode"})
```

```python
postcodes
```

```python
res = paidprices.merge(postcodes,how="left",on="postcode")
```

```python
res = res.merge(postcodes,how="left",left_on="postcode",right_on="pcds")
```

```python
res['date'] = res['date'].apply(lambda x : parse(x) )
```

```python
def convert(dataframe):
    dic = []
    for index, row in dataframe.iterrows():
        item = {
            "postcode":row['postcode'],
            "price":row['price'],
            "geocode":[row['long'],row['lat']],
            "date":row['date']
        }
        dic.append(item)
    return dic
```

```python
jsondoc = convert(res)
pricesdata.insert(jsondoc)
```
